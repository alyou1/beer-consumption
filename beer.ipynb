```python
beer_nameID=df[["beer_name","beer_beerid"]].drop_duplicates()
beerID_by_beer_name=beer_nameID.groupby("beer_name").agg(count=("beer_beerid","count")).reset_index()
beerID_by_beer_name
```

```python
beerID_by_beer_name[beerID_by_beer_name["count"]>1].sort_values(["count"],ascending=False) # Il existe plusieurs noms de bières qui sont liés à plusieurs beer_beerid différents
                                                                                           # On peut penser que plusieurs brasseries fabriquent une meme bière
```

```python
df[["brewery_name","brewery_id"]].nunique()
```

```python
breweryNameId=df[["brewery_name","brewery_id"]].drop_duplicates()
breweryID_by_breweryName=breweryNameId.groupby("brewery_name").agg(count=("brewery_id","count")).reset_index()
breweryID_by_breweryName
```

```python
breweryID_by_breweryName[breweryID_by_breweryName["count"]>1].sort_values(["count"],ascending=False).nunique() # 59 brasseries différentes qui portent le meme nom.
                                                                                                               # Cela peut s'expliquer par le fait qu'une brasserie peut posséder plusieurs filiales
```
